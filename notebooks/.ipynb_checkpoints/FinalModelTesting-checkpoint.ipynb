# Evaluate and test the trained model
#### Extract the optimal threshold from validation data and use to classify both validation and test data
##### optimal threshold = TruePositiveRate-(1-FalsePositiveRate)
#### Evaluate metrics on validation and test data
##### Metrics: ROC_AUC, precision, recall, f1_score, f2_score
#### Plot missclassified images

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
import shutil
import random
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_fscore_support
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, activations
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, BatchNormalization, Dropout, Activation
from keras.regularizers import l1, l2, l1_l2
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import Callback, CSVLogger, ModelCheckpoint
from tensorflow.keras.models import load_model    
from tensorflow.keras.metrics import AUC
sns.set( font_scale = 1.25, style='darkgrid', rc={"lines.linewidth": 2})
```

```python
def ImageGenerators(path_train, path_val, path_test):
    
    train_datagen=ImageDataGenerator(rescale=1./255)
    train_ds=train_datagen.flow_from_directory(
        path_train,  
        class_mode='binary',
        #subset="training", 
        seed=11, 
        target_size=(img_height, img_width), 
        batch_size=batch_size,
        classes=['Non-cracked','Cracked_aug'])

    val_datagen=ImageDataGenerator(rescale=1./255)
    val_ds=val_datagen.flow_from_directory(
        path_val,  
        class_mode='binary',
        #subset="validation", 
        seed=11, 
        target_size=(img_height, img_width), 
        batch_size=batch_size,
        classes=['Non-cracked','Cracked'],
        shuffle=False)

    test_datagen=ImageDataGenerator(rescale=1./255)

    test_ds=test_datagen.flow_from_directory(
        path_test, 
        class_mode='binary',
        seed=11, 
        target_size=(img_height, img_width), 
        batch_size=batch_size,
        classes=['Non-cracked','Cracked'],
        shuffle=False)
    return train_ds, val_ds, test_ds
```

```python
# Image Generators
batch_size = 128
img_height = 256 #224
img_width = 256 #224
root='/storage/Crack_detection/Dataset2/Crack_detection/Dataset2/' #paid machine
pavements_train=os.path.join(root, 'Pavements','Train')
pavements_val=os.path.join(root, 'Pavements','Validation')
pavements_test=os.path.join(root, 'Pavements','Test')
pavements_train, pavements_val, pavements_test=ImageGenerators(pavements_train,pavements_val, pavements_test)
```

```python
def Find_Optimal_Cutoff(target, predicted):
    fpr, tpr, threshold = roc_curve(target, predicted)
    i = np.arange(len(tpr)) 
    roc = pd.DataFrame({'tf' : pd.Series(tpr-(1-fpr), index=i), 'threshold' : pd.Series(threshold, index=i)})
    roc_t = roc.iloc[(roc.tf-0).abs().argsort()[:1]]

    return list(roc_t['threshold']) 

def plot_cm_val_test(val_true,val_pred, test_true, test_pred):
    cm=confusion_matrix(val_true,val_pred)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ['{0:0.0f}'.format(value) for value in cm.flatten()]
    labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
    labels = np.asarray(labels).reshape(2,2)
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    plt.subplots_adjust(wspace=0.5)
    ax=sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
    ax.set(xlabel='Prediction', ylabel='Actual')
    plt.title('Confusion Matrix for Val data')
    
    cm=confusion_matrix(test_true,test_pred)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ['{0:0.0f}'.format(value) for value in cm.flatten()]
    labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
    labels = np.asarray(labels).reshape(2,2)
    plt.subplot(1,2,2)
    ax=sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
    ax.set(xlabel='Prediction', ylabel='Actual')
    plt.title('Confusion Matrix for Test data')
    
    return ax

def plot_cm(true,pred, title):
    cm=confusion_matrix(true,pred)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ['{0:0.0f}'.format(value) for value in cm.flatten()]
    #group_percentages = ['{0:.2%}'.format(value) for value in cm.flatten()/np.sum(cm)]
    #labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
    labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
    labels = np.asarray(labels).reshape(2,2)
    ax=sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
    ax.set(xlabel='Prediction', ylabel='Actual')
    plt.title(title)
    return ax

# evaluate and plot ROC
def plot_roc_val_test(val_y_true, val_score, test_y_true, test_score):
    
    fpr, tpr, threshold=roc_curve(val_y_true,val_score)
    
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    plt.subplots_adjust(wspace=0.5)
    plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % auc(fpr, tpr))
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for Validation data')
    plt.legend(loc='center right')

    # create the axis of thresholds (scores)
    ax2 = plt.gca().twinx()
    ax2.plot(fpr, threshold, markeredgecolor='r',linestyle='dashed', color='r')
    ax2.set_ylabel('Threshold',color='r')
    #ax2.set_ylim([threshold[-1],threshold[0]])
    ax2.set_xlim([fpr[0],fpr[-1]])
    
    fpr, tpr, threshold=roc_curve(test_y_true,test_score)
    plt.subplot(1,2,2)
    plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % auc(fpr, tpr))
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for Test data')
    plt.legend(loc='center right')

    # create the axis of thresholds (scores)
    ax2b = plt.gca().twinx()
    ax2b.plot(fpr, threshold, markeredgecolor='r',linestyle='dashed', color='r')
    ax2b.set_ylabel('Threshold',color='r')
    #ax2.set_ylim([threshold[-1],threshold[0]])
    ax2b.set_xlim([fpr[0],fpr[-1]])
    
    #return ax2

# evaluate and plot ROC
def plot_roc(y_true, score, title):
    fpr, tpr, threshold=roc_curve(y_true,score)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % auc(fpr, tpr))
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc='center right')

    # create the axis of thresholds (scores)
    ax2 = plt.gca().twinx()
    ax2.plot(fpr, threshold, markeredgecolor='r',linestyle='dashed', color='r')
    ax2.set_ylabel('Threshold',color='r')
    #ax2.set_ylim([threshold[-1],threshold[0]])
    ax2.set_xlim([fpr[0],fpr[-1]])
    return ax2


def PlotWrongPredictions(prediction, pavements_test, n_images=50,nx=5,ny=10):
    
    test_pred_classes=[1 if i>0.5 else 0 for i in prediction]
    wrong_prediction_index=[]
    for position, value in enumerate(test_pred_classes):
        if value != pavements_test.classes[position]:
            wrong_prediction_index.append(position)
    
    img_n_rows=n_images//nx+1        
    image_idx=0
    plt.figure(figsize=(nx*4, ny*8))
    for idx in wrong_prediction_index:
        if image_idx<n_images:
            image=pavements_test[int(idx//batch_size)][0][int((idx/batch_size-idx//batch_size)*batch_size)]
            plt.subplot(20, 5, image_idx + 1)
            #plt.subplots(img_n_rows,5, image_idx)
            plt.imshow(image)
            plt.title('Class:' + str(pavements_test.classes[idx]) +'\n'+'Prediction:' + str(np.round(prediction[idx][0],2)))
            plt.xticks([])
            plt.yticks([])
            image_idx+=1    
    plt.show()
```

```python
# load model and predict val/test data
path='/storage/Crack_detection/Dataset2/Crack_detection/opt2/'
model_name='model8b_BN_dp0_dp15_1fcl_l2p0001_021521_lr-decstep_35.h5'

model=load_model(os.path.join(path, model_name))
val_pred=model.predict(pavements_val)
test_pred=model.predict(pavements_test)
```

# Evaluate metrics on validation and test datasets
### Find the optimal threshold from the validation data (defined as the min of TruePositiveRate-(1-FalsePositiveRate)) and use it for validation and test data 

```python
#evaluate on validation and test data
threshold =  np.round(Find_Optimal_Cutoff(pavements_val.classes, val_pred)[0],3)
print('Optimal Threshold =', threshold)

val_pred_classes=[1 if i>threshold else 0 for i in val_pred]
test_pred_classes=[1 if i>threshold else 0 for i in test_pred]
plot_roc_val_test(pavements_val.classes, val_pred, pavements_test.classes, test_pred)
plot_cm_val_test(pavements_val.classes, val_pred_classes, pavements_test.classes, test_pred_classes)
fpr_val, tpr_val, _ =roc_curve(pavements_val.classes, val_pred)
fpr_test, tpr_test, _ =roc_curve(pavements_test.classes, test_pred)
p_val, r_val, f1_val, _ = precision_recall_fscore_support(pavements_val.classes, val_pred_classes, average='binary', beta=1)
p_test, r_test, f1_test, _ = precision_recall_fscore_support(pavements_test.classes, test_pred_classes, average='binary', beta=1)
p_val, r_val, f2_val, _ = precision_recall_fscore_support(pavements_val.classes, val_pred_classes, average='binary', beta=2)
p_test, r_test, f2_test, _ = precision_recall_fscore_support(pavements_test.classes, test_pred_classes, average='binary', beta=2)
metrics_df=pd.DataFrame({'validation':[auc(fpr_val, tpr_val), p_val, r_val, f1_val, f2_val], 'test':[auc(fpr_test, tpr_test), p_test, r_test, f1_test, f2_test]}, index=['ROC_AUC','precision', 'recall', 'f1_score', 'f2_score'])
np.round(metrics_df.T, 3)
```

# Plot false positive examples
